```python
def veri_donustur(klasor_adi):
    goruntuler=dosya(klasor_adi)
    goruntu_sinifli=[]
    
    for goruntu in goruntuler:     # görüntüler listesindeki her bir görüntüyü (dosyanın adresini) tek tek alır.
        goruntu_oku=img.open(goruntu).convert('L')     # adresteki görüntüyü açar, gri tonlamaya (RGB yerine tek renk kanalına
                                                      #(Luminance)) çevirir.  
        goruntu_boyutlandirma=goruntu_oku.resize((28,28))  # görüntüyü 28x28 boyutlarına ölçeklendirerek standart hale getirir.
        goruntu_donusturme=np.array(goruntu_boyutlandirma).flatten()  # görüntüyü (28,28) şeklinde 2D boyutunda bir arraye,
                                              #daha sonra da (784,) uzunluğunda 1D bir array vektöre çevirir.    
        if klasor_adi==gözlüksüz  :  # Gözlüksüz görüntüye 0 sınıf etiketini ekler.
            veriler=np.append(goruntu_donusturme, [0])
        elif klasor_adi==gözlüklü:     # Gözlüklü görüntüye 1 sınıf etiketini ekler.
            veriler=np.append(goruntu_donusturme, [1])
        else:
            continue
        goruntu_sinifli.append(veriler)   # İşlenmiş veriyi listeye ekler ve döndürür.
    
    return goruntu_sinifli
        
```
